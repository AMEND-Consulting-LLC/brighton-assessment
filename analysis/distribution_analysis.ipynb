```python
# Work center processing time distribution plots
    # all values
sns.set(rc={'figure.facecolor':'white'})
g = sns.FacetGrid(cleaned, col="Work Ctr", col_wrap=3, height=4, sharex=False, sharey=False)
g.map(sns.histplot, 'Actual Processing Time', bins=20, kde=True)
g.fig.subplots_adjust(top=0.97)
g.fig.suptitle('ALL OBSERVATIONS')
```

```python
# filter for same day and next day completion
ops_filtered = cleaned.copy()
ops_filtered['same_day'] = np.where(ops_filtered['Act. start'] == ops_filtered['Act.finish'], 1, 0)
ops_filtered['days_to_complete'] = (pd.to_datetime(ops_filtered['Act.finish']) - pd.to_datetime(ops_filtered['Act. start'])).dt.days

# same day
ops_same_day = ops_filtered[ops_filtered['days_to_complete'] == 0]

# next day
ops_next_day = ops_filtered[ops_filtered['days_to_complete'].isin([0,1])]
```

```python
# Work center processing time distribution plots
    # only times completed same day as start date
sns.set(rc={'figure.facecolor':'white'})
g = sns.FacetGrid(ops_same_day, col="Work cntr.", col_wrap=3, height=4, sharex=False, sharey=False)
g.map(sns.histplot, 'Actual Processing Time', bins=20, kde=True)
g.fig.subplots_adjust(top=0.97)
g.fig.suptitle('COMPLETED SAME DAY')
```

```python
# Work center processing time distribution plots
    # only times completed within next day from start date
sns.set(rc={'figure.facecolor':'white'})
g = sns.FacetGrid(ops_next_day, col="Work cntr.", col_wrap=3, height=4, sharex=False, sharey=False)
g.map(sns.histplot, 'Actual Processing Time', bins=20, kde=True)
g.fig.subplots_adjust(top=0.97)
g.fig.suptitle('COMPLETED WITHIN NEXT DAY')
```

```python
# Work center hours per day -- all observations
matplotlib.rcParams.update(matplotlib.rcParamsDefault)

hrs_per_day = cleaned.groupby(by=['Work Ctr', 'Act. start'], as_index=False).sum('Actual Processing Time')
hrs_per_day = hrs_per_day[['Work Ctr', 'Act. start', 'Actual Processing Time']]
hrs_per_day['Actual Processing Time'][hrs_per_day['Actual Processing Time'] > 24] = 24 ### Assign 24 hrs when > 24, for hours spanning multiple days

# compute average by work center, calculate downtime
avg_per_day = hrs_per_day.groupby(by=['Work Ctr'], as_index=False).mean('Actual Processing Time')
avg_per_day['Downtime'] = 24 - avg_per_day['Actual Processing Time']

# sort for plot
avg_per_day = avg_per_day.sort_values(by=['Actual Processing Time'], ascending=True)

# plot values
x = avg_per_day['Work Ctr']
y1 = avg_per_day['Actual Processing Time']
y2 = avg_per_day['Downtime']

# plot size
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)

# plot horixontal bar
plt.barh(x, y1, color='green', label='Operating Time')
plt.barh(x, y2, left=y1, color='red', label='Downtime')
plt.xlim(0,24.1)
plt.title('Avg Daily Runtime by Work Center - ALL OBSERVATIONS')
plt.ylabel('Work Center')
plt.xlabel('Hours')
plt.legend(ncol=2, loc='upper center')

# reference lines
plt.axvline(x = 8, color = 'black', linestyle='dotted')
plt.axvline(x = 16, color = 'black', linestyle='dotted')

plt.text(7.88, -3.7, '8')
plt.text(15.88, -3.7, '16')

plt.show()
```
